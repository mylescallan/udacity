## Class to Download JSONs

```python
import sys
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
import os
import webbrowser, os.path
import datetime
from dateutil import parser
from dateutil import tz
EST = tz.gettz("America/New_York")
import pandas as pd
pd.set_option('display.max_colwidth', None)
import json
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
import matplotlib.ticker as ticker
import numpy as np
from pprint import pprint

import warnings
warnings.filterwarnings('ignore')

__author__ = 'Myles'

os.chdir(output_path)

class MentorReviewStudentFeedbacks(object):
    '''setUp() = Access Udacity's Mentor Review Page
       downloadFeedbacks() = Download feedback information
       downloadReviews() = Download review information
    '''

    def __init__(self,chromedriver_local_path,logon_webpage,email,password):
        # Allow Notifications
        self.opt = Options()
        # Pass the argument 1 to allow and 2 to block
        self.opt.add_experimental_option("prefs", { 
        #     "profile.default_content_setting_values.media_stream_mic": 1, 
        #     "profile.default_content_setting_values.media_stream_camera": 1,
        #     "profile.default_content_setting_values.geolocation": 1, 
            "profile.default_content_setting_values.notifications": 1 
        })
        # Setup
        self.chromedriver_local_path = chromedriver_local_path
        self.logon_webpage = logon_webpage
        self.email = email
        self.password = password
        self.driver = webdriver.Chrome(self.chromedriver_local_path,options=self.opt)
        self.driver.set_window_position(0, 0)
        self.driver.set_window_size(1466, 1124)

    def setUp(self):
        # quick sign-in
        self.driver.get(self.logon_webpage)
        self.email_elem = WebDriverWait(self.driver, 15).until(EC.presence_of_element_located((By.CSS_SELECTOR, "input[id='email']")))
        self.email_elem.send_keys(self.email)
        self.password_elem = WebDriverWait(self.driver, 15).until(EC.presence_of_element_located((By.CSS_SELECTOR, "input[id='revealable-password']")))
        self.password_elem.send_keys(self.password + Keys.ENTER)
        time.sleep(2)
        
    def flatten_json(self,y):
        '''If JSON is hierarchical, flatten
           Source: https://towardsdatascience.com/flattening-json-objects-in-python-f5343c794b10
        '''
        out = {}
        def flatten(x, name=''):
            if type(x) is dict:
                for a in x:
                    flatten(x[a], name + a + '_')
            elif type(x) is list:
                i = 0
                for a in x:
                    flatten(a, name + str(i) + '_')
                    i += 1
            else:
                out[name[:-1]] = x

        flatten(y)
        return out   
    
    def downloadJSON(self,webpage,date,filename,download=False):
        myDate = lambda x: '{}'.format(x.strftime("%m/%d/%y at %I:%M%p"))
        time_date = datetime.datetime.now().strftime("%H_%M%_on_%b_%d")
        # access review content page
        self.driver.get(webpage)
        content = self.driver.find_element_by_xpath("html").text
        content_json = json.loads(content)
        content_list = [self.flatten_json(d) for d in content_json]
        revenue = {}
        for d in content_list:
            d[date] = parser.parse(d[date]).astimezone(EST)
            d['Date'] = myDate(d[date])
        content_list = sorted(content_list, key=lambda d: d[date])
        for d in content_list:
            if 'price' in list(d.keys()):
                if not revenue.get(d[date].strftime("%b-%Y")):
                    revenue[d[date].strftime("%b-%Y")] = round(float(d['price']),2)
                else:
                    revenue[d[date].strftime("%b-%Y")] += round(float(d['price']),2)
                d['Earnings'] = revenue[d[date].strftime("%b-%Y")]
        content_list = sorted(content_list, key=lambda d: d[date], reverse=True)
        if download:
            with open(filename.format(time_date), 'w') as f:
                f.write(json.dumps(content_list,sort_keys=True, default=str))
        return content_list
    
    def tearDown(self):
        self.driver.quit()
        
```

```python
def find_nth(haystack, needle, n):
    '''find nth occurance of a substring in a string'''
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
```

```python
class CreateReviewsWebpage(object):
    '''createTable() = creates a HTML table
       createContents() = creates HTML page (contents will be HTML table)
       strToFile() = write the HTML page to file
       browseLocal() = open the created HTML file in browser
    '''
    def __init__(self,reviewList,fileName,title = "Feedback Page", collapse =[]):
        '''Required inputs: A list of lists, a filename, and a page name'''
        self.reviewList = reviewList
        self.fileName = fileName
        self.title = title
        self.collapse = collapse
        self.new = 0
        
    def createTable(self, listOfLists):
        """Create the HTML table, from the list of lists generated from the JSON (or other methods)"""
        htmlTable = "<table id='pyTable'><tr>"
        # loop over list of lists (which are JSONs for each review)
        for ind, rows in enumerate(listOfLists):
            for col, data in enumerate(rows):
                # add either the table headers (list 0) or the table rows (all other list items)
                if ind == 0:
                    htmlTable += "<th>{}</th>".format(data)
                else:
                    if col in self.collapse:
                        if find_nth(data, '</h1>', 1) > 0:
                            strIndex = find_nth(data, '</h1>', 1) + 6
                        elif find_nth(data, '</p>', 1) > 0:
                            strIndex = find_nth(data, '</p>', 1) + 5
                        elif len(data) > 200:
                            strIndex = len(data.split('. ',2)[0]) + 2
                        else:
                            strIndex = len(data)
                        htmlTable += "<td><button class='collapsible'>{}</button><div class='content'>{}</div></td>".format(data[:strIndex],data[strIndex:])
                    else:
                        htmlTable += "<td>{}</td>".format(data)
            
            htmlTable += "</tr><tr>".format(data)
        # remove the last <tr>, close the table
        htmlTable = htmlTable[:-4] + "</table>"
        return htmlTable
    
    def createContents(self):
        '''Create the HTML page. 
        The text will be the HTML table generated from createTable and the data'''
        createdTable = self.createTable(self.reviewList)
        contents = """<!DOCTYPE html PUBLIC '-//W3C//DTD HTML 4.01 Transitional//EN'>
            <html>
            <head>
              <meta content='text/html; charset=ISO-8859-1'
             http-equiv='content-type'>
              <title>{0}</title>
            </head>
            <style>
            body {{
            width:100vw !important;
            }}
            table {{
            width:100vw !important;
            }}
            th {{
            width: 5vw !important;
            }}
            
            .collapsible {{
              background-color: #777;
              color: white;
              cursor: pointer;
              padding: 18px;
              width: {1}vw !important;
              border: none;
              text-align: left;
              outline: none;
              font-size: 15px;
            }}

            .active, .collapsible:hover {{
              background-color: #555;
            }}

            .collapsible:after {{
              content: '\\002B';
              color: white;
              font-weight: bold;
              float: right;
              margin-left: 5px;
            }}

            .active:after {{
              content: '\\2212';
            }}

            .content {{
              padding: 0 18px;
              max-height: 0;
              width: {1}vw !important;
              overflow: hidden;
              transition: max-height 0.2s ease-out;
              background-color: #f1f1f1;
            }}
            #pyTable {{
                font-family: 'Trebuchet MS', Arial, Helvetica, sans-serif;
                border-collapse: collapse;
                width: 100%;
            }}
            #pyTable td, #pyTable th {{
                border: 1px solid #ddd;
                padding: 8px;
            }}
            #pyTable tr:nth-child(even){{background-color: #f2f2f2;}}
            #pyTable tr:hover {{background-color: #ddd;}}
            #pyTable th {{
                padding-top: 12px;
                padding-bottom: 12px;
                text-align: left;
                background-color: #4CAF50;
                color: white;
            }}
            </style>
            <body>{2}
            <script>
            var coll = document.getElementsByClassName("collapsible");
            var i;

            for (i = 0; i < coll.length; i++) {{
              coll[i].addEventListener("click", function() {{
                this.classList.toggle("active");
                var content = this.nextElementSibling;
                if (content.style.maxHeight){{
                  content.style.maxHeight = null;
                }} else {{
                  content.style.maxHeight = content.scrollHeight + "px";
                }} 
              }});
            }}
            </script>
            </body>
            </html>""".format(self.title, ((100-(5*len(self.reviewList[0])))/len(self.collapse))-1, createdTable)
        return contents

    def strToFile(self):
        """Create the HTML page. Then write the HTML file with the given name and the given contents."""
        self.contents = self.createContents()
        output = open(self.fileName,"w")
        output.write(self.contents)
        output.close()

    def browseLocal(self):
        '''Open chrome webbrowser with the local file created above
        with the given filename.'''
        import webbrowser, os.path
        self.strToFile()
        if not self.fileName.startswith("Reviews_"):
            # Note: remove .get(using='chrome') if you are using your default browser to log on to udacity (and it is not chrome)
            webbrowser.get().open("file:///" + os.path.abspath(self.fileName),new=self.new) #elaborated for Mac
            self.new = 2
        
```

```python
import markdown
md = markdown.Markdown()

def mdToHTML(aList, keyList):
    '''Convert values with keys in keyList
       if empty, set to No Entry'''
    # for each dictionary in the list
    for d in aList:
        # for each key in the keyList
        for cols in keyList:
            # if the value exists and is a string
            if d[cols] and isinstance(d[cols],str):
                # convert Markdown to HTML
                d[cols] = md.convert(d[cols])
            # otherwise
            else:
                # change the value to "No Entry"
                d[cols] = "No Entry"
```

```python
#  All arguments set as defaults (change values above)
import markdown
from collections import defaultdict

        

def callCreateReviewsWebpage(aList, byProject = True, project = 'project_name', linkID = linkID, pageTitle = pageTitle, sortBy=sortBy, sortOrder = sortOrder ,
                             includeCols = includeCols, defaultHeaders = False, optHeaders = optHeaders, fileName=fileName,
                            convertMarkdown = convertMarkdown, markdownCols = markdownCols, collapse=collapse):
    
    
    ## create instance of markdown
    md = markdown.Markdown()

    ## Create link (from submission number)
    for d in aList:
        d['Link'] = "<a href=https://review.udacity.com/#!/reviews/{0} target='_blank'>{0}</a>".format(d[linkID])

    ## Columns to include in HTML table
    dHTML = [{k: d[k] for k in includeCols} for d in aList]
    
    if byProject:
        # create separate lists for dictionaries depending on project name
        dd = defaultdict(list)
        for d in dHTML:
            dd[(d[project])].append(d)

        ## Create separate pages for each project
        for item in list(dd.values()):

            ## create a list of lists of entries
            myList = [list(d.values()) for d in item]    

            ## add the column headers (will become table headers)
            if defaultHeaders:
                myList.insert(0,list(item[0].keys()))
            else:
                ## or add your own headers
                myList.insert(0,optHeaders)

            webpage = CreateReviewsWebpage(myList,fileName=fileName.format(item[0][project].replace("/","_")),title=pageTitle, collapse=collapse)
            webpage.browseLocal()
    else:
        ## create a list of lists of entries            
        myList = [list(d.values()) for d in dHTML]    

        ## add the column headers (will become table headers)
        if defaultHeaders:
            myList.insert(0,list(list(item[0].keys())))
        else:
            ## or add your own headers
            myList.insert(0,optHeaders)

        webpage = CreateReviewsWebpage(myList,fileName=fileName, title=pageTitle, collapse=collapse)
        webpage.browseLocal()      
```

```python
# Determine the number of Reviews that were not evaluated
## Print the count and percentage
no_comments = ratings_df[ratings_df.body.str.lower().str.contains('no entry')]
print('''The number of reviews with ratings: {},
the number of reviews with comments {}, {:.2f}%'''. format(len(ratings_df), len(ratings_df) - len(no_comments),
                                                           (1 - (len(no_comments)/len(ratings_df)))*100))


# Synonyms
synonymsMain = {"most": ["most"],
                "best": ["best"],
                "best_synonyms": ["best", "excellent", "awesome", "wonderful", "perfect", "superb",
                         "incredible", "impressive", "outstanding", "most",
                         "extremely helpful", "insightful", "fantastic"],
                "excellent": ["best", "great", "excellent", "awesome", "wonderful", "perfect", "appreciate", "superb",
                           "incredible", "impressive", "outstanding", "extremely helpful", "insightful", "fantastic"],
                "good": ["best", "great", "excellent", "awesome", "wonderful", "perfect", "appreciate", "superb",
                      "incredible", "impressive", "outstanding", "extremely helpful", "insightful",  "fantastic",
                      "detailed", "helpful", "helped", "thank you", "thanks", "learn", "very good"],
               "empty": []}

def searchEvals(df, col, synonyms, keyword, rating=5, text = True):
    '''input: dataframe, column, synonyms dictionary, keyword
       output: print out of the percentage of reviews, and reviews with ratings, that contain one of the synonyms'''
    # list of synonyms to search for
    searchFor = synonyms[keyword]
    # entries that contain 'No Entry' (i.e. that have no evaluation)
    no_comments = df[df[col].str.lower().str.contains('no entry')]
    # all ratings
    numRatings = len(df)
    # limit rating values
    df = df[df.rating >= rating].copy()
    # search the evaluations for the words
    search = df[df[col].str.lower().str.contains('|'.join(searchFor))]
    # mean pass rate
    pass_rate = np.mean(search.result == "passed")*100   
    if text:
        print("\nThe list used:", ", ".join(searchFor) )
        # print a breakdown of the results
        print('''    Evaluations limited to those with a rating of {} or above,
        {:.2f}% pass rate with {} evaluations,
        {}: The number of reviews with ratings,
        {}: The number of reviews with written evaluations,
        {}: The number of reviews with {} evaluations,
        {:.2f}% of all rated reviews,
        {:.2f}% of all evaluations with comments'''. format(rating,
                                                            pass_rate, keyword,
                                                            numRatings,
                                                            numRatings- len(no_comments),
                                                            len(search), keyword,
                                                            (len(search)/numRatings)*100,
                                                            (len(search)/(numRatings- len(no_comments)))*100))
    return search

# Best synonyms
best = searchEvals(ratings_df2, 'body', synonymsMain, 'best')

# Best synonyms
best_synonyms = searchEvals(ratings_df2, 'body', synonymsMain, 'best_synonyms')

# Excellent synonyms
excellent = searchEvals(ratings_df2, 'body', synonymsMain, 'excellent')

# Excellent synonyms
good = searchEvals(ratings_df2, 'body', synonymsMain, 'good')
```

```python
# Calculate the Average (and Count) of Ratings, for each project, for each 60 day window

results = {}
# groupby each project
for ind, grp in combined_df.groupby(['project_name']):
    # date of earliest rated review for each project
    min_date = grp.Date.min() + pd.DateOffset(days=60) ## 2 month window
#     min_date = min_date.tz_localize('US/Eastern')
    # today
    date = grp.Date.max()
    date = pd.to_datetime('today')
    date = date.tz_localize('US/Eastern')

    temp_period = []
    # loop between earliest and latest review dates
    while date >= min_date:
        # 60 day window from 'date'
        period = date - pd.DateOffset(days=60) ## 2 month window
        # ratings for the 60 day window
        period_ratings = grp[(grp.Date >= period) & (grp.Date <=  date)].rating
        # average
        period_ratings_mean = period_ratings.apply(pd.to_numeric, errors='coerce').mean()
        # count
        period_ratings_count = period_ratings.apply(pd.to_numeric, errors='coerce').count()
        # store as a dictionary
        temp = {"date": date, "short_date": date.strftime("%d %B, %Y"), "mean": period_ratings_mean, "count": period_ratings_count}
        # save to list
        temp_period.append(temp)
        # scroll forward one day
        date -= pd.DateOffset(days=1)
    # for each project, store results in dictionary
    results[ind] = temp_period
    

```

```python
def rating_visualization(df_dailyRatings):
    '''60 day window visualization of projects'''
    for plot_type in ["mean", "count"]:
    
        def adjust(val):
            if plot_type == "mean":
                return round(val,2)
            else:
                return int(val)

        fig, ax = plt.subplots(figsize=[14.70, 8.27])
        switch = -3
        textp = 0
        for item in df_dailyRatings.project.unique():
            grp = df_dailyRatings[df_dailyRatings.project==item].copy()
            if not grp.empty:
                grp.plot(x = "date", y = plot_type, ax = ax, label = item, color = color[item], rot = 0)
                # 60 day average annotations
                grp.iloc[::28, :][["mdates", "mean", "count"]].apply(lambda row: ax.annotate(adjust(row[plot_type]),(row["mdates"],row[plot_type]),
                                                                                             size=12, ha='center', va='bottom', color = "red",
                                                                                             xytext=(25, 15*switch),
                                                                                             textcoords='offset points', arrowprops=dict(arrowstyle='-|>'),
                                                                                             bbox = {"boxstyle":"circle,pad=0.3", "fc":colorbbox[item],
                                                                                                     "ec":color[item], "lw":2}),
                                                                     axis=1);
                # Mean Line
                ax.axhline(y=grp[plot_type].mean(), color=color[item], linestyle='-.')
                # Latest 60 day average
                ax.annotate("Average:\n"+str(round(grp[plot_type].mean(),2)),(grp["mdates"].max(),grp[plot_type].mean()),
                            size=12, ha='center', va='bottom', weight = "bold", color = "white",
                            xytext=(1.2, 1 - textp), textcoords='axes fraction',
                            arrowprops=dict(arrowstyle='-|>'),bbox = {"boxstyle":"circle,pad=0.3", "fc":color[item],
                                                                                                        "ec":colorbbox[item], "lw":2})
                # Overall averages
                ax.annotate("Latest:\n{}\n({})".format(str(round(grp.iloc[0,:]["mean"].mean(),2)), str(grp.iloc[0,:]["count"])),
                            (grp.iloc[0,:]["mdates"], grp.iloc[0,:]["mean"]),
                            size=12, ha='center',
                            va='bottom', weight = "bold",
                            color = "white",
                                xytext=(1 - textp, 1.02), textcoords='axes fraction',
                            bbox=dict(facecolor = color[item], edgecolor=colorbbox[item], boxstyle='round,pad=0.25'))
                switch = 3
                textp += .15

        ax.set_title("Average Ratings", fontsize = 12, weight = "bold")    
        ax.set_ylabel("Average Ratings", fontsize = 10, weight = "bold")    
        ax.set_xlabel(" ")    
        plt.suptitle("60 day rolling window - {}".format(plot_type).title(), y = 1.04, fontsize = 14, weight = "bold")    
        first_legend = plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.4),
                   title="Project", title_fontsize = 12,
                                 handlelength=2);

        # Add the legend manually to the current Axes.
        ax_leg = plt.gca().add_artist(first_legend)

        # Create another legend for the second line.
        from matplotlib.lines import Line2D

        red_circle = Line2D([0], [0], marker='o', color='w', label='Count              ',
                                markerfacecolor='lightblue', markeredgecolor='blue', linewidth = 2, markersize=15),

        plt.legend(handles=red_circle, loc='center left', bbox_to_anchor=(1.05, 0.15),
                                 handlelength=2);
        
        gap = (df_dailyRatings[plot_type].max() - df_dailyRatings[plot_type].min()) * .2
        ylim_height = df_dailyRatings[plot_type].max() + gap
        ylim_lower = df_dailyRatings[plot_type].min() - gap
        plt.ylim(ylim_lower,ylim_height)

        # Create a duplicate of the original xaxis, giving you an additional axis object
        ax1 = ax.twinx()
        # Set the limits of the new axis from the original axis limits
        ax1.set_ylim(ax.get_ylim())
        plt.tight_layout()
        plt.show()
```
